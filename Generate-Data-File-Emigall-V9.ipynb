```python
def createINITrecord(EXT_ID, 
                     INIT_BPKIND, 
                     INIT_BU_GROUP, 
                     INIT_BU_RLTYP, 
                     INIT_BU_TYPE,
                     INIT_PARTNER):
    # **********************************
    # xdf1[x][0] --> EXT_ID
    # xdf1[x][1] --> INIT_BPKIND
    # xdf1[x][2] --> INIT_BU_GROUP
    # xdf1[x][3] --> INIT_BU_RLTYP
    # xdf1[x][4] --> INIT_BU_TYPE
    # xdf1[x][5] --> INIT_PARTNER
    # **********************************

    if INIT_PARTNER is None:
        INIT_PARTNER = ''
    if INIT_BPKIND is None:
        INIT_BPKIND = ''
    if INIT_BU_RLTYP is None:
        INIT_BU_RLTYP = ''
    if INIT_BU_GROUP is None:
        INIT_BU_GROUP = ''
    
    v_INIT_record_formatted = f'{EXT_ID}\tINIT\t{INIT_PARTNER}\t{INIT_BU_RLTYP}\t{INIT_BU_TYPE}\t{INIT_BU_GROUP}\t{INIT_BPKIND}'
    
    return v_INIT_record_formatted
```

```python
# --------------------------------------------------------------------------
# Determine number of to be generated based on Number of record in each file 
# and actual number of records in excel sheet
# --------------------------------------------------------------------------
def determineNumberOfFiles(i_numberOfRecordsInEachFile, i_lastcell):
    o_numberOfFiles2Generate = (i_lastcell % i_numberOfRecordsInEachFile)
    l_quotient  = i_lastcell // i_numberOfRecordsInEachFile
    l_mod = i_lastcell % i_numberOfRecordsInEachFile
    if(l_mod > 0):
        o_numberOfFiles2Generate = l_quotient + 1
    else: 
        o_numberOfFiles2Generate = l_quotient
    return o_numberOfFiles2Generate
```

```python
numberOfFiles2Generate = determineNumberOfFiles(numberOfRecordsInEachFile, lastcell)
print('Total Number of files to be generated: {}'.format(numberOfFiles2Generate)) 
```

```python
def extractDataChunkForFile(fromCount, toCount, iterationCount):
    wb = None
    app = None
    data_sheet = None
    df1 = None
    
    wb = xw.Book(r'DATA-Emigall\WIP-Emigall-BP-Source-Data-SMALLv2.xlsx') 
    # Start invisible Excel
    app = xw.App(visible=False, add_book=False)
    data_sheet = wb.sheets['DATATAB_1']
    # Reading from Excel as Dataframe
    if (int(iterationCount) == 1 and skipHeaderRecord):
        df1 = data_sheet.range((fromCount+1,1),(toCount,40)).value
    else:    
        df1 = data_sheet.range((fromCount,1),(toCount,40)).value
    wb.close()
    return df1
```

```python
# ***********************************************************************
# The parameter and it's description
# fileName - Tells what should be the file name for the program to create
# fromCount - It's the record number where it should start from in the input 
#             file (in this case it will be the Excel row)
# toCount - record number where it should stop of that iteration
# ***********************************************************************
def createDataFileIfNotExists(fileName, iterationCount):
    
    # Calculate fromCount & toCount based on the iteration number
    
    fromCount = None
    toCount = None
    
    if (numberOfRecordsInEachFile >= lastcell):
        fromCount = 1
        toCount = lastcell
    else:
        if (iterationCount == 1):
            fromCount = 1
            toCount = numberOfRecordsInEachFile
        elif (iterationCount == numberOfFiles2Generate):
            fromCount = (numberOfRecordsInEachFile * (iterationCount - 1)) + 1
            toCount = lastcell
        else:
            fromCount = (numberOfRecordsInEachFile * (iterationCount - 1) + 1)
            toCount = (fromCount + numberOfRecordsInEachFile) -1
            
    print('The current count block is: {} & {} and I am in iteration {} & number of file to generate is: {}'.format(fromCount,toCount,iterationCount,numberOfFiles2Generate)) 
    
    
    f_content_new_line = "\n"
    with open(fileName,'w') as f:
        
        xdf1 = extractDataChunkForFile(fromCount,toCount,iterationCount)
        print(len(xdf1))
        # looping
        for x in range(0, len(xdf1), 1):
            # **********************************
            # xdf1[x][0] --> EXT_ID
            # xdf1[x][1] --> INIT_BPKIND
            # xdf1[x][2] --> INIT_BU_GROUP
            # xdf1[x][3] --> INIT_BU_RLTYP
            # xdf1[x][4] --> INIT_BU_TYPE
            # xdf1[x][5] --> INIT_PARTNER
            # **********************************
            f_content_INIT = createINITrecord(xdf1[x][0], 
                                              xdf1[x][1], 
                                              xdf1[x][2], 
                                              xdf1[x][3], 
                                              xdf1[x][4], 
                                              xdf1[x][5])
            f.write(str(f_content_INIT))
            f.write(f_content_new_line)
            
            # **********************************
            # xdf1[x][0]   --> EXT_ID
            # xdf1[x][6]   --> BUT000_AUGRP
            # xdf1[x][36]  --> BUT000_BU_SORT1
            # xdf1[x][37]  --> BUT000_BU_SORT2
            # xdf1[x][7]   --> BUT000_NAME_ORG1
            # xdf1[x][8]   --> BUT000_NAME_ORG2
            # xdf1[x][9]   --> BUT000_NAME_ORG3
            # xdf1[x][10]  --> BUT000_NAME_ORG4
            # xdf1[x][11]  --> BUT000_SOURCE
            # xdf1[x][12]  --> BUT000_XBLCK
            # xdf1[x][38]  --> BUT000_XDELE
            # **********************************
            
            f_content_BUT000 = createBUT000record(xdf1[x][0], 
                                                  xdf1[x][6], 
                                                  xdf1[x][36], 
                                                  xdf1[x][37], 
                                                  xdf1[x][7], 
                                                  xdf1[x][8],
                                                  xdf1[x][9],
                                                  xdf1[x][10],
                                                  xdf1[x][11],
                                                  xdf1[x][12],
                                                  xdf1[x][38])
            f.write(str(f_content_BUT000))
            f.write(f_content_new_line)

            # **********************************
            # xdf1[x][0]   --> EXT_ID
            # xdf1[x][13]  --> BUT020_CHIND_ADDR
            # xdf1[x][39]  --> BUT020_NAME_CO
            # xdf1[x][14]  --> BUT020_CITY1
            # xdf1[x][15]  --> BUT020_COUNTRY
            # xdf1[x][16]  --> BUT020_LANGU
            # xdf1[x][17]  --> BUT020_LOCATION
            # xdf1[x][18]  --> BUT020_POST_CODE1
            # xdf1[x][19]  --> BUT020_PO_BOX
            # xdf1[x][20]  --> BUT020_REGION
            # xdf1[x][21]  --> BUT020_SMTP_ADDR
            # xdf1[x][22]  --> BUT020_STREET
            # xdf1[x][23]  --> BUT020_STR_SUPPL1
            # xdf1[x][24]  --> BUT020_STR_SUPPL2
            # xdf1[x][25]  --> BUT020_STR_SUPPL3
            # xdf1[x][26]  --> BUT020_TAXJURCODE
            # xdf1[x][27]  --> BUT020_TEL_NUMBER
            # **********************************
            
            # Make changes to routine which formats the string and not here ....

            f_content_BUT020 = createBUT020record(xdf1[x][0], 
                                                  xdf1[x][13],
                                                  xdf1[x][39],
                                                  xdf1[x][14], 
                                                  xdf1[x][15],
                                                  xdf1[x][16], 
                                                  xdf1[x][17], 
                                                  xdf1[x][18], 
                                                  xdf1[x][19], 
                                                  xdf1[x][20], 
                                                  xdf1[x][21], 
                                                  xdf1[x][22], 
                                                  xdf1[x][23], 
                                                  xdf1[x][24], 
                                                  xdf1[x][25], 
                                                  xdf1[x][26], 
                                                  xdf1[x][27])
            f.write(str(f_content_BUT020))
            f.write(f_content_new_line)
            
            f_content_TAXNUM = createTAXNUMrecord(xdf1[x][0], 
                                                  xdf1[x][28], 
                                                  xdf1[x][29], 
                                                  xdf1[x][30])
            f.write(str(f_content_TAXNUM))
            f.write(f_content_new_line)

            f_content_BUT0ID = createBUT0IDrecord(xdf1[x][0], 
                                                  xdf1[x][31], 
                                                  xdf1[x][32], 
                                                  xdf1[x][33], 
                                                  xdf1[x][34], 
                                                  xdf1[x][35])
            f.write(str(f_content_BUT0ID))
            f.write(f_content_new_line)
            
            f_content_ENDE = createENDLINErecord(xdf1[x][0])
            f.write(str(f_content_ENDE))
            f.write(f_content_new_line)
        return f.mode
```
